```python
import requests
from bs4 import BeautifulSoup 
import json
import csv
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import random
import pandas as pd
```

```python
#***DO NOT RUN THIS CELL***
#This cell is to extract the urls but I have already run this cell and saved the urls in the url_cleansers variable
url_cleansers = []


for page_num in range(1, 6):
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    # Set up the webdriver
    driver = webdriver.Chrome()
    driver.get(f"https://www.sephora.com/ca/en/shop/face-wash-facial-cleanser?currentPage={page_num}")

    # Scroll to the end of the page
    elem = driver.find_element(By.TAG_NAME, "body")
    no_of_pagedowns = 20
    while no_of_pagedowns:
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(random.uniform(0.5, 1.5))
        no_of_pagedowns -= 1

    # Get the HTML content of the page
    html = driver.page_source

    # Close the webdriver
    driver.close()

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    if soup.select(".css-foh208"):
        products1 = soup.select(".css-foh208")
        for j in range(0,len(products1)):
            prod = products1[j]
            link = prod.select("a")[0]['href']
            url_cleansers.append(link)

    if soup.select(".css-1qe8tjm"):
        products2 = soup.select(".css-1qe8tjm")
        for j in range(0,len(products2)):
            prod = products2[j]
            link = prod.select("a")[0]['href']
            url_cleansers.append(link)

    time.sleep(random.uniform(1.0, 3.0))
```

```python
data_toners = []
chrome_options = Options()
chrome_options.add_argument('--headless')


for url in url_toners:
   
    try:
        driver = webdriver.Chrome()
        driver.get(url)
        elem = driver.find_element(By.TAG_NAME, "body")
        no_of_pagedowns = 20
        while no_of_pagedowns:
            elem.send_keys(Keys.PAGE_DOWN)
            time.sleep(random.uniform(0.5, 1.5))
            no_of_pagedowns -= 1
        html = driver.page_source
        driver.close()
        soup = BeautifulSoup(html, 'html.parser')

        try:
            name = soup.select(".css-11zrkxf")
            product_name = name[0].find("a").text
            brand_name = name[0].find("span").text
        except:
            product_name = ""
            brand_name = ""

        try:
            cost = soup.select(".css-1tzthcm")
            cost = cost[0].find("b").text
        except:
            cost = ""

        try:
            size = soup.select(".css-1ag3xrp")
            size = size[0].find("span").text
        except:
            size = ""

        script = soup.find('script', {'id': 'linkStore'})
        script_content = script.contents[0]
        script_json = json.loads(script_content)

        if 'highlights' in script_json['page']['product']['currentSku']:
            highlights = script_json['page']['product']['currentSku']['highlights']
            highlight_tag = [highlight['altText'] for highlight in highlights]
            highlight_tag_str = '; '.join(highlight_tag)
        else:
            highlight_tag_str = ""

        if 'longDescription' in script_json['page']['product']['productDetails']:  
            long_des = script_json['page']['product']['productDetails']['longDescription']
        else: 
            long_des = ""
        if 'rating' in script_json['page']['product']['productDetails']:
            rating = script_json['page']['product']['productDetails']['rating']
        else:
            rating = ""
        if 'productId' in script_json['page']['product']['productDetails']:
            product_id = script_json['page']['product']['productDetails']['productId']
        else:
            product_id = ""
        if 'lovesCount' in script_json['page']['product']['productDetails']:
            loves_count = script_json['page']['product']['productDetails']['lovesCount']
        else: 
            loves_count = ""
        if 'reviews' in script_json['page']['product']['productDetails']:
            reviews = script_json['page']['product']['productDetails']['reviews']
        else: 
            reviews = ""
        if 'ingredientDesc' in script_json['page']['product']['currentSku']:
            ingredients = script_json['page']['product']['currentSku']['ingredientDesc']
        else: 
            ingredients = ""


        data_toners.append([product_name, brand_name,cost,size, highlight_tag_str, long_des, rating, product_id, loves_count, reviews, ingredients,url])
        time.sleep(random.uniform(1.0, 3.0))
    except:
        pass
```
