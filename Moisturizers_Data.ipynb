```python
data_moisturizers = []
#there are 772 links in moisturizers 
url_100 = url_all[701:773]
chrome_options = Options()
chrome_options.add_argument('--headless')


for url in url_100:
   
    try:
        driver = webdriver.Chrome()
        driver.get(url)
        elem = driver.find_element(By.TAG_NAME, "body")
        no_of_pagedowns = 20
        while no_of_pagedowns:
            elem.send_keys(Keys.PAGE_DOWN)
            time.sleep(random.uniform(0.5, 1.5))
            no_of_pagedowns -= 1
        html = driver.page_source
        driver.close()
        soup = BeautifulSoup(html, 'html.parser')

        try:
            name = soup.select(".css-11zrkxf")
            product_name = name[0].find("a").text
            brand_name = name[0].find("span").text
        except:
            product_name = ""
            brand_name = ""

        try:
            cost = soup.select(".css-1tzthcm")
            cost = cost[0].find("b").text
        except:
            cost = ""

        try:
            size = soup.select(".css-1ag3xrp")
            size = size[0].find("span").text
        except:
            size = ""

        script = soup.find('script', {'id': 'linkStore'})
        script_content = script.contents[0]
        script_json = json.loads(script_content)

        if 'highlights' in script_json['page']['product']['currentSku']:
            highlights = script_json['page']['product']['currentSku']['highlights']
            highlight_tag = [highlight['altText'] for highlight in highlights]
            highlight_tag_str = '; '.join(highlight_tag)
        else:
            highlight_tag_str = ""

        if 'longDescription' in script_json['page']['product']['productDetails']:  
            long_des = script_json['page']['product']['productDetails']['longDescription']
        else: 
            long_des = ""
        if 'rating' in script_json['page']['product']['productDetails']:
            rating = script_json['page']['product']['productDetails']['rating']
        else:
            rating = ""
        if 'productId' in script_json['page']['product']['productDetails']:
            product_id = script_json['page']['product']['productDetails']['productId']
        else:
            product_id = ""
        if 'lovesCount' in script_json['page']['product']['productDetails']:
            loves_count = script_json['page']['product']['productDetails']['lovesCount']
        else: 
            loves_count = ""
        if 'reviews' in script_json['page']['product']['productDetails']:
            reviews = script_json['page']['product']['productDetails']['reviews']
        else: 
            reviews = ""
        if 'ingredientDesc' in script_json['page']['product']['currentSku']:
            ingredients = script_json['page']['product']['currentSku']['ingredientDesc']
        else: 
            ingredients = ""


        data_moisturizers.append([product_name, brand_name,cost,size, highlight_tag_str, long_des, rating, product_id, loves_count, reviews, ingredients,url])
        time.sleep(random.uniform(1.0, 3.0))
    except:
        pass
        
    


```
