```python
import pandas as pd
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# I define the stop_words here so I don't do it every time in the function below
stop_words = stopwords.words('english')
```

```python
def get_keywords2(row):
    some_text = row['Skincare Concern']
    if isinstance(some_text, str):  # add this line to check if the value is a string
        lowered = some_text.lower()
        tokens = nltk.tokenize.word_tokenize(lowered)
        keywords = [keyword for keyword in tokens if keyword.isalpha() and not keyword in stop_words]
        keywords_string = ','.join(keywords)
        return keywords_string
    else:
        return ''  # or any other default value you want to return if the value is not a string
```
